```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

we need first 3 dictionaries and obtain name

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head(3)
```

```python
def stem1(text):
    y=[]
    for i in text.split():
        ps.stem(i)
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
# we need to find 
# we will not find eculedian distance, the more the dimention the les reliable eculedian distance becomes
# we will find cosine distance between the 2 vectors which is angle between the two vectors, the less the theeta (angle) the more similar the movies are
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[1] # diagnoal will always be 1
```

```python
# index of that movie to obtain that similarity row
# sort those similarity 
# fetch top 5
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0] # fetching index
    distances = similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    #return movies_list
```
